```python
import pandas as pd
```

```python
data = pd.read_csv("influencer.csv")
```

```python
def to_num(x):
    
    x = x.strip()
    if(x[-1] == "億"):
        return float(x[:-1])*100000000
    elif(x[-1] == "萬"):
        return float(x[:-1])*10000
    else:
        return x
```

```python
data.head()
```

```python
data.fans_cnt = data.fans_cnt.apply(to_num)
data.view_cnt = data.view_cnt.apply(to_num)
```

```python
data.head()
```

```python
result = data.groupby("category").agg({"fans_cnt":"mean"}).reset_index()
```

```python
result
```

```python
result.category.tolist()
```

```python
import seaborn as sns
sns.set_theme(style="darkgrid")
ax = sns.barplot(x="category", y="fans_cnt", data=result)
```

```python
result = data.groupby("category").agg({"view_cnt":"mean"}).reset_index()
```

```python
import seaborn as sns
sns.set_theme(style="darkgrid")
ax = sns.barplot(x="category", y="view_cnt", data=result)
```
